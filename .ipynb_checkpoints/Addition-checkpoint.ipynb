## Binary Addition

We use two models for this problem: Multi-Layer Perceptron and Convolution Neural Network. To visualize the information plane, considering the limited computation power, we choose the MLP approach and compute encoder/decoer mututal information based on a fraction of training data. 

### Model 1: Multilayer perceptron w/ Batch Normalization

```python
from utils import load_mnist
import utils
import numpy as np
from mlp_vec import Layer, LayerArgs, Model, ModelArgs
from collections import Counter
import math
from mi_tool import MI
import plot_tool
#reload(mlp_vec)
```

```python
%matplotlib inline
```

```python
from random import randint
print randint(0, 32767) 
```

```python
def make_bit_array(num, length):
    bi = bin(num)[2:]
    remain = length -len(bi)
    return [int(_) for _ in list("0"*remain + bi)]
```

```python
print bin(32767)
print len(bin(65535+65535))-2
```

```python
def create_instance(total):
    for _ in xrange(total):
        num1 = randint(0, 32767) 
        num2 = randint(0, 32767) 
        summed = num1+ num2
        feat = []
        feat += make_bit_array(num1, 16)
        feat += make_bit_array(num2, 16)
        res = []
        res += make_bit_array(summed, 17)
        yield feat, res
```

```python
X, y = [], []
```

```python
total = 10000
train_num = int(total * 0.8)
```

```python
train_num 
```

```python
for feat, res in create_instance(total):
    X.append(feat)
    y.append(res)
```

```python
X, y = np.array(X), np.array(y)
X_train, y_train = X[:train_num], y[:train_num]
X_test, y_test = X[train_num:], y[train_num:]
```

```python
# define the network structure with 2 hidden layers of dimension 100 and 20. 
# other parameters were set to default, activation function is set to RELU by default. 

layer_args = [LayerArgs(32, 32, layer_type = "INPUT"), \
              LayerArgs(32, 100), \
              LayerArgs(100, 50), \
              LayerArgs(50, 17, layer_type = "OUTPUT", activate = np.exp)]

# our model is using mini-batch gradient descent
# set max #(epoch) as 80 and max #(iteration) as 100000;
# For everay 1000 iterations, we compute (and plot) MI;
model_args = ModelArgs(num_passes = 8000, max_iter=10000000, report_interval=1500)

# intialize the model
# TODO: model weights intialization by denoised autoencoder

model = Model(layer_args, model_args)
model.feed_data(X_train, y_train, X_test, y_test)
model.trial_data(X_train, y_train)
model.intialize_model()
```

```python
MI_client = MI(X_train, y_train, 10)
MI_client.discretize()
MI_client.pre_compute()
```

```python
for epoch, hidden_layers in model.run_model():
    MI_client.mi_single_epoch(hidden_layers, epoch)
```
